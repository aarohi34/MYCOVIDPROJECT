```python
msk = np.random.rand(len(file)) < 0.8

train = file[msk]
test = file[~msk]
```

```python
plt.scatter(train.CONFIRMED, train.Recovered,  color='blue')
plt.xlabel("trainconfiremed")
plt.ylabel("trainrecoveed")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['CONFIRMED']])  #numpy.asanyarray()function is used when we want to convert input to an array
train_y = np.asanyarray(train[['Recovered']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Telengana.csv')
```

```python
df
```

```python
df.head()
```

```python

plt.figure(figsize=(8,7))
plt.bar(df.Date,df.Confirmed,label='Confirmed cases',color='blue')
plt.title("covid_analysis_june_july:telengana")
plt.xticks(df.Date,rotation=90)
plt.xlabel("CASES DAILY (1stjune-15july)",labelpad=10)
plt.ylabel("Number of Cases ",labelpad=10)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.bar(df.Date,df.Confirmed,label='Confirmed cases',color='blue')
plt.bar(df.Date,df.Death ,color='green',label='Death')
plt.bar(df.Date,df.Recovered,label='recovered',color='red')
plt.legend()
plt.title("covid_analysis_june_july_telengana")
plt.xticks(df.Date,rotation=90)
plt.xlabel("FROM 1STJUNE-15JULY",labelpad=10)
plt.ylabel("no of cases",labelpad=10)
plt.show()
```

```python

from matplotlib import style
style.use('ggplot')

plt.figure(figsize=(9,4))
plt.plot(df.Date,df.Confirmed,label='Confirmed Cases',color='blue',linewidth=5)
plt.plot(df.Date,df.Death,color='green',label='Death',linewidth=5)
plt.plot(df.Date,df.Recovered,label='Recovered',color='red')

plt.legend()
plt.title("covid_analysis_june_july_telengana")

plt.xticks(rotation=90)
plt.xlabel("Daily cases",labelpad=20)
plt.ylabel("Number of cases",labelpad=5)
plt.grid(True,color='k')

plt.show()
```

```python
IFR_DATEWISE=(df['Recovered']/df['Death'])*100
```

```python
plt.figure(figsize=(8, 6))
#plt.bar(df_filtered.Date,df_hosp.Beds,color="grey",label='Total Samples')
plt.bar(df.Date,IFR_DATEWISE,label='Confirmed Cases',color='lightskyblue')
plt.bar(df.Date,df.Death,color="black",label='Death Cases')
plt.title("COVID-19 TELENGANA_IFR")
plt.xticks(df.Date,rotation=90)
plt.xlabel("Date",labelpad=10)
plt.ylabel("IFR",labelpad=10)
plt.show()
```
